```python
from keras_preprocessing.image import ImageDataGenerator
```

```python
def data_generator_asmat(TRAINING_DIR, VALIDATION_DIR, TESTING_DIR):
    train_datagen = ImageDataGenerator(rescale=1./255.,
                                     rotation_range=50,
                                     width_shift_range=0.2,
                                     height_shift_range=0.2,
                                     shear_range=0.2,
                                     zoom_range=0.2,
                                     horizontal_flip=True,
                                     fill_mode='nearest')
    validation_datagen = ImageDataGenerator(rescale=1./255,
                                            rotation_range=50,
                                            horizontal_flip=True)
    testing_datagen = ImageDataGenerator(rescale=1./255,
                                         rotation_range=50,
                                         horizontal_flip=True)


    train_generator_asmat = train_datagen.flow_from_directory(directory=TRAINING_DIR,
                                                      batch_size=16,
                                                      class_mode='categorical',
                                                      target_size=(150, 150),
                                                      save_to_dir="dataset_model/train/asmat",
                                                      save_prefix='model_',
                                                      save_format='png',)
    

    validation_generator_asmat = validation_datagen.flow_from_directory(directory=VALIDATION_DIR,
                                                      batch_size=8,
                                                      class_mode='categorical',
                                                      target_size=(150, 150),
                                                      save_to_dir="dataset_model/val/asmat",
                                                      save_prefix='model_',
                                                      save_format='png',)


    testing_generator_asmat = testing_datagen.flow_from_directory(directory=TESTING_DIR,
                                                      batch_size=8,
                                                      class_mode='categorical',
                                                      target_size=(150, 150),
                                                      save_to_dir="dataset_model/test/asmat",
                                                      save_prefix='model_',
                                                      save_format='png',)
    
    for i in range(87):
        train_generator_asmat.next()
    for i in range(24):
        validation_generator_asmat.next()
    for i in range(12):
        testing_generator_asmat.next()
```
